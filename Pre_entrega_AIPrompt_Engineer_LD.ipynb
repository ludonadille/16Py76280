```python
# Texto de entrada

# Aquí debería incluir los distintos inputs

servicio = input("Ingrese el servicio: ")
sub_servicio = input("Ingrese el sub-servicio: ")
ftes = input("Ingrese la cantidad de FTEs: ")
cola = input("Ingrese la cláusula de COLA: ")
pago = input("Ingrese los términos de pago: ")
cant_pais = input("Ingrese la cantidad de países: ")
paises =[]
for i in range(int(cant_pais)):
    pais = input("Ingrese el país: ")
    paises.append(pais)
```

```python
# Prompt con formato esperado
prompt = f"""
En función de las siguientes características del servicio: {servicio} , {sub_servicio}, cantidad de FTEs {ftes} y
condiciones comerciales indicadas por el Cliente tales como:
tiene cláusula de COLA? {cola},
los términos de pago: {pago} y
los países en los que se presta el servicio: {paises}.
Necesito que me indiques: 1) Los competidores que pueden
estar presentes en esta contienda, 2) los precios por hora o FTE anual, y 3) sus fortalezas y debilidades de los competidores con presencia en estos servicios.
Además necesito que sea en formato JSON (sin ´´´json para poder integrar en otro sistema).
Siguien la siguiente estructura clave-valor (esta será un combinación de los inputs del prompt y del output del prompt):
'servicio', 'sub-servicio', 'competidor', 'cantidad de FTEs', 'claúsula de COLA', 'términos de pago', 'cantidad de países', 'país' (según la cantidad seleccionada),
'precio por hora', 'precio por FTE anual' , 'fortalezas' , 'debilidades'.
"""
```

## Llamada básica a Gemini

```python
model = genai.GenerativeModel("gemini-2.5-flash")

contenido = model.generate_content(prompt).text
print(contenido)
```

```python
# Guardar el JSON en un archivo

with open('analisis_comp.json', 'w') as f:
    f.write(contenido)
```
